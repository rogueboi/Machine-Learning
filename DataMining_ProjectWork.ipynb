# **3.** CLASSIFICATION

Classification is the task of learning a target function $f$ that maps each attribute set x to one of the predefined class labels $y$. The target function is also known informally as a **classification model**. A classification model is useful for the following purposes - 
1. **Descriptive Modeling :** A classification model can serve as an explanatory tool to distinguish between objects of different classes.
2. **Predictive Modeling :** A classification model can also be used to predict the class label of unknown records. A classification model can be treated as a black box that automatically assigns a class label when presented with the attribute set of an unknown record.  

Classification techniques are most suited for predicting or describing data sets with binary or nominal categories. They are less effective for ordinal
categories because they do not consider the implicit order among the categories.Other forms of relationships, such as the subclass–superclass relationships among categories are also ignored.

## **3.1** Naive Bayes' Classification

#### **1. Conditional Probability :**  

Conditional Probability of event $E$ given that event $F$ has already occured is defined as : 
$$ P(E|F) = \frac{Number\space of\space elementary\space events\space favourable\space to\space E\cap F}{Number\space of\space elementary\space events\space favourable\space to\space F}  
= \frac{n(E\cap F)}{n(F)}  
= \Large\frac{\frac{n(E\cap F)}{n(S)}}{\frac{n(F)}{n(S)}}
\normalsize= \frac{P(E\cap F)}{P(F)} $$

Hence,

$$ P(E|F) P(F) = P(E\cap F) $$  
$$ and $$  
$$ P(F|E) P(E) = P(F\cap E) = P(E\cap F) $$  

$$\Large\therefore\space \normalsize P(E|F) = \frac{P(F|E)\space P(E)}{P(F)} $$

---
#### **2. Theorem of Total Probability :**  

Let $ {E_{1}, E_{2},...,E_{n}} $ be a partition of the sample space S, and suppose that each of the events $ E_{1}, E_{2},...,E_{n} $ has nonzero probability of occurrence.  
Let $A$ be any event associated with $S$, then  
$$ P(A) = P(E_{1})\space P(A|E_{1}) + P(E_{2})\space P(A|E_{2}) + ... + P(E_{n})\space P(A|E_{n})  
= \Large\sum_{j=1}^{n} \normalsize P(E_{j}) P(A|E_{j}) $$

---
#### **3. Bayes’ Theorem :**  

According to **Bayes' Theorem**, if $ E_{1}, E_{2},...,E_{n} $ are n non empty events which constitute a partition of sample space $S$, i.e. $ E_{1}, E_{2},...,E_{n} $ are pairwise disjoint and $ E_{1}\cup E_{2}\cup...\cup E_{n} = S $ and $F$ is any event of nonzero probability, then
$$ P(E|F) = \frac{P(E∩F)}{P(F)}  
= \frac{P(F|E)\space P(E)}{P(F)}  
= \frac{P(F|E)\space P(E)}{\Large\sum_{j=1}^{n} \normalsize P(E_{j})\space P(F|E_{j})} \;\;\;\;\;\; \normalsize\forall i = 1, 2, 3, ..., n $$  

**Bayes' Theorem in the language of Machine Learning :** 

In machine learning we are often interested in determining the best hypothesis from some space $H$, given the observed training data $D$.  
One way to specify, the concept of best hypothesis is to say that we demand the most probable hypothesis, given the data D plus any initial knowledge about the prior probabilities of the various hypotheses in $H$.  

_Bayes' theorem provides a way to calculate the probability of a hypothesis based on its prior probability, the probabilities of observing various data given the hypothesis, and the observed data itself._
$$ P(h|D) = \frac{P(D|h)\space P(h)}{P(D)} $$

1. $ P(h|D) $ is called the **posterior probability of** $h$, because it reflects our confidence that $h$ holds after we have seen the training data $D$.
2. $ P(D|h) $ is often referred to as **likelihood of the data** $D$ given $h$.
3. $ P(h) $ is often called the **prior probability of** $h$ and may reflect any background knowledge we have about the chance that $h$ is a correct hypothesis.
4. $ P(D) $ is often called the **marginal probability of** $D$.

---
#### **4. Naive Bayes' Classifier :**  

**_A Naive Bayes' Classifier assumes that the presence of a particular feature in a class is unrelated to the presence of any other feature._**

* In many learning scenarios, the learner considers some set of candidate hypotheses $H$ and is interested in finding the most probable hypothesis $h \in H $ given the observed data $D$ (or at least one of the maximally probable if there are several). Any such maximally probable hypothesis is called a **maximum a posteriori (MAP) hypothesis**.  

* We can determine the **MAP hypothesis** by using **Bayes' theorem** to calculate the posterior probability of each candidate hypothesis. More precisely, we will say that $ h_{MAP}$ is a _MAP hypothesis_ provided  

$$ h_{MAP} = \max_{h \in H} \normalsize P(h|D)  
= \max_{h \in H} \normalsize\frac{P(h|D)\space P(h)}{P(D)}  
= \max_{h \in H} \normalsize P(h|D)\space P(h) $$

> **Notice in the final step above we dropped the term $P(D)$ because it is a constant independent of $h$.**  

* As we know, $P(D|h)$ is called the **likelihood of the data** $D$ given $h$, and any hypothesis that maximizes $P(D|h)$ is called a **maximum likelihood (ML)** hypothesis,  

$$ h_{ML} = \max_{h \in H} \normalsize P(D|h) $$  
* The fundamental Naive Bayes' assumption is that each feature makes an:
    - **independent** &
    - **equal**  
contribution to the outcome. Therefore,  
$$ h_{ML} = \max_{h \in H} \normalsize P(D|h)  
= \max_{h \in H} \Large\prod_{i=1}^{n} \normalsize P(d_{i}|h) $$
$$ where\space n = no.\space of\space features\space in\space the\space data\space instance\space D $$  
* The Naive Bayes' classifier applies to learning tasks where each instance $x$ is described by a conjunction of attribute values and where the target function $f(x)$ can take on any value from some finite set $V$. 
* A set of training examples of the target function is provided, and a new instance is presented, described by the tuple of attribute values $[a_{1}, a_{2},...a_{n}]$. The learner is asked to predict the target value, or classification, for this new instance.  
* The Bayesian approach to classifying the new instance is to assign the most probable target value, $v_{MAP}$ given the attribute values $[a_{1}, a_{2},...a_{n}]$ that describe the instance.
$$ v_{MAP} = \max_{v_{j}\in V} P(v_{j}|[a_{1}, a_{2},...a_{n}])$$  
* We can use Bayes' theorem to rewrite this expression as  
$$ v_{MAP} = \max_{v_{j}\in V} \frac{P([a_{1}, a_{2},...a_{n}]|v_{j})\space P(v_{j})}{P([a_{1}, a_{2},...a_{n}])}  
= \max_{v_{j}\in V} P([a_{1}, a_{2},...a_{n}]|v_{j})\space P(v_{j})$$  
&emsp;&emsp; Since, the event for each feature $a_{i}$ to be some $x_{i}$ is independent of each other, then, according to the theorem of Total Probability,  
$$ P([a_{1}, a_{2},...a_{n}]) = \Large\sum_{j=1}^{m} \normalsize P(v_{j}) P([a_{1}, a_{2},...a_{n}]|v_{j})  
= \Large\sum_{j=1}^{m}( \normalsize P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) \Large) $$  
$$where\space m = no.\space of\space distinct\space classes\space in\space the\space outcome$$

**Naive Bayes' Classification**  
$$ \max_{v_{j}\in V}  P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) $$
$  \Large\therefore \;\;\normalsize The\; actual\; conditional\; probability\; of\; the\; target\; value\; is\; given\; as-\; $
$$ \max_{v_{j}\in V}  \frac{P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j})}{\Large\sum_{j=1}^{m}( \normalsize P(v_{j}) \Large\prod_{i=1}^{n} \normalsize P(a_{i}|v_{j}) \Large)}$$  

---
#### **5. Estimating Probabilities :**  

If the training dataset is such that one of the conditional probabilities is zero, then the entire expression for that particular outcome becomes zero. In order to avoid such scenarios we can adopt a Bayesian approach to estimate the probability, using the m-estimate defined as follows-  
$$ m-estimate: P(x_{i} = c|v_{j}) = \frac{n_{c} + m\space p}{n + m} $$
$$ where\space n_{c} = no.\space of\space times\space the\space event\space is\space observed\space to\space occur $$
$$ n = total\; no.\; of\; oppurtunities $$
$$ p = \frac{1}{no.\; of\; possible\; values\; of\; the\; attribute\; x_{i}} $$
$$ m = equivalent\; sample\; size\; (for\; laplace\; smoothing\; m = no.\; of\; possible\; values\; of\; the\; attribute\; x_{i}) $$

## **3.2** Decision Tree Classification

#### **1. Introduction**

We can solve a classification problem by asking a series of carefully crafted questions about the attributes of the test record. Each time we receive an answer, a follow-up question is asked until we reach a conclusion about the class label of the record. The series of questions and their possible answers can be organized in the form of a decision tree, which is a hierarchical structure consisting of nodes and directed edges.  

The tree has three types of nodes:
* A **root node** that has no incoming edges and zero or more outgoing edges.
* **Internal nodes**, each of which has exactly one incoming edge and two or more outgoing edges.
* **Leaf or terminal nodes**, each of which has exactly one incoming edge and no outgoing edges.

In a decision tree, each **leaf node** is assigned a class label. The **non-terminal nodes**, which include the root and other internal nodes, contain attribute test conditions to separate records that have different characteristics.


---
#### **2. Hunt's Algorithm**

In Hunt’s algorithm, a decision tree is grown in a recursive fashion by partitioning the training records into successively purer subsets. Let $D_{t}$ be the set of training records that are associated with node $t$ and $y = {y_{1}, y_{2}, . . . , y_{c}}$ be the class labels. The following is a recursive definition of Hunt’s algorithm.  
* **Step 1:** If all the records in $D_{t}$ belong to the same class $y_{t}$, then $t$ is a leaf node labeled as $y_{t}$.  
* **Step 2:** If $D_{t}$ contains records that belong to more than one class, an attribute test condition is selected to partition the records into smaller subsets. A child node is created for each outcome of the test condition and the records in $D_{t}$ are distributed to the children based on the outcomes. The algorithm is then recursively applied to each child node.

---
#### **3. Methods for Expressing Attribute Test Conditions**

Decision tree induction algorithms must provide a method for expressing an attribute test condition and its corresponding outcomes for different attribute types.  
* **Binary Attributes :** The test condition for a binary attribute generates two potential outcomes.
* **Nominal Attributes :** Test condition can be expressed in two ways,
    * For a multiway split, the number of outcomes depends on the number of distinct values for the corresponding attribute.
    * For binary splits by considering all $2^{k-1}$ ways of creating a binary partition of $k$ attribute values.
* **Ordinal Attributes :** Ordinal attribute values can be grouped as long as the grouping does not violate the order property of the attribute values.
* **Continuous Attributes :** For continuous attributes, the test condition can be expressed as a comparison test $(A < v)$ or $(A > v)$ with binary outcomes, or a range query with outcomes of the form $v_{i} < A < v_{i+1},\; \forall i=1,...,k$. For the binary case, the decision tree algorithm must consider all possible split positions v, and it selects the one that produces the best partition. For the multiway split, the algorithm must consider all possible ranges of continuous values.


---
#### **4. Measures for Selecting the Best Split**

**ENTROPY :** Entropy is a measure of randomness (overall purity) of a given dataset. The mathematical formula for Entropy is:
$$ Entropy(t) = -\sum_{j=1}^{n} p(i|t)\log_{2}(p(i|t)) $$  
The Higher the Entropy, the lower the purity of the split. The graph of Entropy for binary classification ($+$ & $-$) is as follows :

<img src="https://miro.medium.com/max/1130/1*M15RZMSk8nGEyOnD8haF-A.png" style="display:block; margin-left:auto; margin-right:auto;"/>

**GINI INDEX :** The mathematical formula for Gini Index is:
$$ Gini(t) = 1 - \sum_{j=1}^{n} [p(i|t)]^2$$  
The Higher the Gini, the lower the purity of the split.

**CLASSIFICATION ERROR :** The mathematical formula for Classification Error is:
$$ Classification\; Error(t) = 1 - \max_{i}[p(i|t)]$$

$$where,\; \mathbf{n} = total\; no.\; of\; distinct\; class\; labels$$
$$\mathbf{p(i|t)} = the\; fraction\; of\; records\; belonging\; to\; class\; i\; at\; a\; given\; node\; t.$$  

**GAIN :** The gain, $\Delta$, is a criterion that can be used to determine the goodness of a split:
$$\Delta = I_{parent} - \sum_{j=1}^{k} \frac{N(v_{j})}{N} I(v_{j})$$

$$where,\; \mathbf{I(*)} = the\; impurity\; measure\; of\; a\; given\; node$$
$$\mathbf{N} = the\; total\; number\; of\; records\; at\; the\; parent\; node$$
$$\mathbf{k} = the\; number\; of\; attribute\; values$$
$$\mathbf{N(v_{j})} = the\; number\; of\; records\; associated\; with\; the\; child\; node,\; v_{j}.$$  

> **When Entropy is used as the impurity measure, the difference in Entropy is known as Information Gain, $\Delta_{info}$**.  

**GAIN RATIO :** Gain Ratio is a strategy to modify the splitting criterion to take into account the number of outcomes produced by the attribute test condition and is given by :
$$ Gain\; Ratio = \frac{\Delta_{info}}{Split\; Info}$$
$$where,\; \mathbf{Split\; Info} = -\sum_{i=1}^{k}P(v_{i})\log_{2}P(v_{i})$$
$$\mathbf{k} = the\; total\; number\; of\; splits$$

## **3.3** K-Nearest Neighbor Classification

#### **1. Introduction**
A nearest neighbor classifier represents each example as a data point in a $d$-dimensional space, where $d$ is the number of attributes. Given a test example, the model computes the proximity of the example to the rest of the data points in the training set, using one of the proximity measures.  

<img alt="Map1NNReducedDataSet" src="https://upload.wikimedia.org/wikipedia/commons/e/e9/Map1NNReducedDataSet.png" style="height:50%; width:50%; display:block; margin-left:auto; margin-right:auto;">

---
#### **2. Classification Technique**
The k-nearest neighbors of a given example $z$ refer to the $k$ points that are closest to $z$. The data point is classified based on the class labels of its neighbors. In the case where the neighbors have more than one label, the data point is assigned to the majority class of its nearest neighbors.  

If $k$ is too small, then the nearest-neighbor classifier may be susceptible to overfitting because of noise in the training data. On the other hand, if
is too large, the nearest-neighbor classifier may misclassify the test $k$ instance because its list of nearest neighbors may include data points that are located far away from its neighborhood.

# **4.** REGRESSION  

Regression analysis is a statistical method performed to estimate the level effect of an independent variable $x$ on a dependent variable $y$. In other words, it is used to understand or describe the relationships between a set of independent variables and dependent variables.  

The outcome of regression analysis is a mathematical equation often called a regression equation.
$$y = \alpha + \beta_{0} x_{0} + \beta_{1} x_{1} + \beta_{2} x_{2} + ...$$  
$\alpha$ and $\beta_{i}$ in the above equations are parameters and they remain constant as $x$ and $y$ changes. “$\alpha$” is the intercept and “$\beta_{i}$” is the slope (for linear regression curves). By determining the values of “$\alpha$” and “$\beta_{i}$” one can calculate the value of “$y$” for a given value of “$x$”.  

Regression analysis is a predictive modelling technique, used to analyse the cause and effect. It is primarily used for:
* Prediction and Forecasting
* Inferring relationships between the independent and dependent variables.

## **4.1** Linear Regression

### **1. Introduction**
Linear Regression is a regression analysis of dependent and independent variables when they exhibit a linear relationship. It is a supervised learning technique and it assumes that the dependence of $y$ on $x_{1}, x_{2},..., x_{p}$ is linear.

---
### **2. Practical Uses**
Linear regression has many practical uses. Most applications fall into one of the following two broad categories:
* If the goal is prediction, linear regression can be used to fit a predictive model to an observed data set of values of the response and explanatory variables. After developing such a model, if additional values of the explanatory variables are collected without an accompanying response value, the fitted model can be used to make a prediction of the response.
* If the goal is to explain variation in the dependent variable that can be attributed to variation in the independent variables, linear regression analysis can be applied to quantify the strength of the relationship between the response and the explanatory variables.

## **4.2** Polynomial Regression 

### **1. Introduction**  

Polynomial regression is the nonlinear relationship between an independent variable $x$ and a dependent variable $y$, and it is modeled as $n^{th}$ degree polynomial in $x$.  

In other words, It is the form of linear regression. In every linear regression equation to find the value of dependent variable(y) the model first needs to find the values of coefficient $\beta_{1}, \beta_{2},..., \beta_{n}$ and intercept $\alpha$. For any values of independent variable $x$, values of coefficient and intercept will remain same.

### **2. Assumptions of Polynomial Regression**  

* The behavior of a dependent variable can be explained by a linear, or curvilinear, additive relationship between the dependent variable and a set of $k$ independent variables $(x_{i},\; i=1\; to\; k)$.
* The relationship between the dependent variable and any independent variable is linear or curvilinear (specifically polynomial).
* The independent variables are independent of each other.
* The errors are independent, normally distributed with mean zero and a constant variance (OLS).

##### **3.2.2 Correlation Matrix**
Correlation coefficients are indicators of the strength of linear relationship between two different variables, $X$ and $Y$.

##### **4.2.1 Discretizing _'Price'_ Column**

The following code discretizes the _Price_ column into roughly 9 equal sized bins (categories) namely **Low Tier-1**, **Low Tier-2**, **Low Tier-3**, **Medium Tier-1**, **Medium Tier-2**, **Medium Tier-3**, **High Tier-1**, **High Tier-2**,  & **High Tier-3**.

````python
binLabels = ['Low Tier-1', 'Low Tier-2', 'Low Tier-3', 'Medium Tier-1', 'Medium Tier-2', 'Medium Tier-3', 'High Tier-1', 'High Tier-2', 'High Tier-3']
data['Price Class'] = pd.qcut(data['Price'], q=9, labels=binLabels)
````

````python
TypeGrouped = pd.DataFrame(data['Type'].groupby(data['Price Class']).value_counts())
TypeGrouped.columns = ['Counts']
TypeGrouped.reset_index(inplace=True)
TypeGroupMax = pd.merge(TypeGrouped, TypeGrouped.groupby('Price Class', as_index=False).agg({'Counts' : 'max'}), on=['Price Class', 'Counts'], how='inner')
TypeGroupMax.columns = ['Price Class', 'Type', 'Max Count']
TypeGroupMax
````

A confusion matrix summarizes the number of instances predicted correctly or incorrectly by a classification model.

The following terminology is often used when referring to the counts tabulated in a confusion matrix:
* **True positive (TP) or $f_{++}$**, which corresponds to the number of positive examples correctly predicted by the classification model.
* **False negative (FN) or $f_{+-}$**, which corresponds to the number of positive  examples wrongly predicted as negative by the classification model.
* **False positive (FP) or $f_{-+}$**, which corresponds to the number of negative examples wrongly predicted as positive by the classification model.
* **True negative (TN) or $f_{--}$**, which corresponds to the number of negative examples correctly predicted by the classification model.  

**ACCURACY :**  
Accuracy is the measure of number of correct predictions out of total predictions and is given in terms of the elements of the confusion matrix as :
$$ Accuracy\; (a) = \frac{Number\; of\; correct\; predictions}{Total\; number\; of\; predictions} = \frac{f_{++} + f_{--}}{f_{++} + f_{+-} + f_{-+} + f_{--}}$$

**PRECISION :**    
Precision determines the fraction of records that actually turns out to be positive in the group the classifier has declared as a positive class. The higher the precision is, the lower the number of false positive errors committed by the classifier.
$$ Precision\; (p) = \frac{TP}{TP+FP}$$

**RECALL :**  
Recall measures the fraction of positive examples correctly predicted by the classifier. Classifiers with large recall have very few positive examples
misclassified as the negative class. In fact, the value of recall is equivalent to the true positive rate.
$$ Recall\; (r) = \frac{TP}{TP+FN}$$

**$F_{1}$ SCORE :**  
Precision and recall can be summarized into another metric known as the $F_{1}$ measure. In principle, $F_{1}$ Score represents a harmonic mean between recall and precision, i.e.,
$$ F_{1}\; Score\; (F_{1}) = \frac{2}{\frac{1}{r}+\frac{1}{p}}$$
$$Hence,\; F_{1} = \frac{2rp}{r+p} = \frac{2 \times TP}{2 \times TP + FP + FN}$$

* LIBRARY : **_sklearn_**
* MODULE : **_metrics_**
* FUNCTION : __*confusion_matrix()*__, __*accuracy_score()*__, __*precision_score()*__, __*recall_score()*__ & __*f1_score()*__

````python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm1 = pd.DataFrame(confusion_matrix(yC_test, yC_pred1))
cm1.columns.name = 'Predicted Class'
cm1.index.name = 'Actual Class'
print("\nThe Confusion Matrix is :")
display(cm1)

accuracy1 =  accuracy_score(yC_test, yC_pred1)
print("Accuracy = {a} %".format(a = round(accuracy1, 4) * 100))

precision1 = precision_score(yC_test, yC_pred1, average='macro')
print("Precision = {a} %".format(a =  round(precision1, 4) * 100))

recall1 = recall_score(yC_test, yC_pred1, average='macro')
print("Recall = {a} %".format(a =  round(recall1, 4) * 100))

f1_1 = f1_score(yC_test, yC_pred1, average='macro')
print("f1 score = {a}".format(a =  round(f1_1, 4)))
````

**Evaluation Metrics** 

**MEAN SQUARED ERROR (MSE) :**  
In statistics, the mean squared error (MSE) of an estimator (of a procedure for estimating an unobserved quantity) measures the average of the squares of the errors — that is, the average squared difference between the estimated values and what is estimated.  

Mathematical Formula:  
$$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_{i,\; pred} - y_{i})^2$$
where, 
* $\mathbf{n}$ = total number of test set observations
* $\mathbf{y_{i, pred}}$ = predicted value of $i^{th}$ test set observation
* $\mathbf{y_{i}}$ = actual value of $i^{th}$ test set observation 

MSE is a risk function, corresponding to the expected value of the squared error loss. The fact that MSE is almost always strictly positive (and not zero) is because of randomness or because the estimator does not account for information that could produce a more accurate estimate.  

**MEAN ABSOLUTE ERROR (MAE) :**  
The Mean absolute error represents the average of the absolute difference between the actual and predicted values in the dataset. It measures the average of the residuals in the dataset.  

Mathematical Formula:  
$$MAE = \frac{1}{n} \sum_{i=1}^{n} |y_{i,\; pred} - y_{mean}|$$
where, 
* $\mathbf{n}$ = total number of test set observations
* $\mathbf{y_{i,\; pred}}$ = predicted value of $i^{th}$ test set observation
* $\mathbf{y_{mean}}$ = mean value of the test set observations 

**$R^{2}$ SCORE :**  
Coefficient of determination also known as $R^{2}$ score is used to evaluate the performance of a linear regression model. It is the amount of the variation in the output dependent attribute which is predictable from the input independent variable(s). It is used to check how well-observed results are reproduced by the model, depending on the ratio of total deviation of results described by the model.

Mathematical Formula:
$$R^2\; Score= 1- \frac{SS_{res}}{SS_{tot}}$$
where,
* $\mathbf{SS_{res}}$ = sum of squares of the residual errors = $ \sum_{i=1}^{n} (y_{i} - y_{mean})^2$  
* $\mathbf{SS_{tot}}$ = total sum of squares of the errors = $ \sum_{i=1}^{n} (y_{i} - y_{i,\; pred})^2$
* $\mathbf{n}$ = total number of test set observations

<a href="https://www.geeksforgeeks.org/python-coefficient-of-determination-r2-score/">Reference: Python – Coefficient of Determination-R2 score</a>

* LIBRARY : **_sklearn_**
* MODULE : **_metrics_**
* FUNCTION : __*mean_squared_error()*__, __*mean_absolute_error()*__ & __*r2_score()*__

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse1 = mean_squared_error(yR_test, yR_pred1)
mae1 = mean_absolute_error(yR_test, yR_pred1)
r2score1 = r2_score(yR_test, yR_pred1)

print("Mean Squared Error =", mse1)
print("Mean Absolute Error =", mae1)
print("R2 Score =", r2score1)
````

##### **5.2.3 Gradient Boosting**

In gradient boosting algorithm, each predictor corrects its predecessor’s error. The weights of the training instances remains the same for every predictor while, each predictor is trained using the residual errors of predecessor as labels.

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20200721214745/gradientboosting.PNG"  style="height:50%; width:50%; display: block; margin-left: auto;margin-right: auto;"/>  

The ensemble consists of $N$ trees. $Tree\; 1$ is trained using the feature matrix $X$ and the labels $Y$. The predictions labelled $y_{1}$ (hat) are used to determine the training set residual errors $r_{1}$. $Tree\; 2$ is then trained using the feature matrix $X$ and the residual errors $r_{1}$ of $Tree\; 1$ as labels. The predicted results $r_{1}$ (hat) are then used to determine the residual $r_{2}$. The process is repeated until all the $N$ trees forming the ensemble are trained.

<a href="https://www.geeksforgeeks.org/ml-gradient-boosting/">Reference: ML – Gradient Boosting</a>

**Training the Gradient Boosting Model**
* LIBRARY : **_sklearn_**
* MODULE : **_ensemble_**
* CLASS : **_GradientBoostingRegressor_**

````python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(n_estimators=800, subsample=0.1, criterion='squared_error', max_features='auto', min_samples_leaf=5, min_samples_split=5, max_leaf_nodes=40, learning_rate=0.008, max_depth=20, random_state=2, n_iter_no_change=5)
gbr.fit(xR_train, yR_train)
````

**Predicting the Test Set Results**

````python
yR_pred3 = gbr.predict(xR_test)
print('Y -> PREDICTED')
yR_pred3
````

**Evaluation Metrics**

````python
mse3 = mean_squared_error(yR_test, yR_pred3)
mae3 = mean_absolute_error(yR_test, yR_pred3)
r2score3 = r2_score(yR_test, yR_pred3)

print("Mean Squared Error =", mse3)
print("Mean Absolute Error =", mae3)
print("R2 Score =", r2score3)
````
